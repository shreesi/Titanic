```python
# Importing libraries

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
# reading the data

data = pd.read_csv('train.csv')
```

```python
#printig first five rows

data.head()
```

```python
# It shows how many rows and columns does our dataset have

data.shape
```

```python
data.describe()
```

```python
# It shows the data types of the columns and the null values

data.info()
```

```python
# removing unnecessary columns

data.drop(['Name', 'Ticket', 'Cabin'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.info()
```

```python
# Transforming the categorical columns into numerical column

le = LabelEncoder()

for i in range(len(data[['Sex']])):
    data[['Sex']] = le.fit_transform(data[['Sex']])
    
for i in range(len(data[['Embarked']])):
    data[['Embarked']] = le.fit_transform(data[['Embarked']])
```

```python
data.head()
```

```python
# Checking null values

data.isna().sum()
```

```python
# fill the null values with mean

data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data[['Age']]
```

```python
# splitting the independent and dependent dataset
# x = independent dataset
# y = dependent dataset

x = data.drop(['Survived'], axis = 1)
y = data[['Survived']]
```

```python
x
```

```python
y
```

```python
# splitting the dataset into training and testing dataset
# we have splitted 25% dataset fro testing and 75% of dataset into training

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=4)
```

```python
model = SVC()
```

```python
# We have fitted the dataset with model

model.fit(x_train, y_train)
```

```python
# Checking the score of our trained model

model.score(x_test, y_test)
```

```python
# Applying the Standard Scalar to improve the accuracy of our model

sc = StandardScaler()
```

```python
x_train1 = sc.fit_transform(x_train)
x_test1 = sc.fit_transform(x_test)
```

```python
model1 = SVC()
```

```python
model1.fit(x_train1, y_train)
```

```python
model1.score(x_test1, y_test)
```
